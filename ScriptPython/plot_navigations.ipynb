```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_collection_navigation = pd.read_csv("dataset/collection_navigations.csv")
```

```python
df_collection_navigation.head()
```

```python
# Distribution des ventes par heures 
# df_collection_navigation
plt.figure(figsize=(10, 6))
plt.hist(df_collection_navigation['hour_visited'], bins=12, color="#D64550")
plt.xlabel('Heure de la journée')
plt.ylabel('Nombre de visites')
plt.title('Distribution des visites par heure')
plt.grid(False)
plt.show()
```

```python
# Créer un diagramme à barres pour le type d'action
# Créer un diagramme à barres pour le type d'action
plt.figure(figsize=(10, 6))
bar_plot = df_collection_navigation['type_action'].value_counts().plot(kind='bar', color='#D64550')
plt.xlabel('Type d\'action')
plt.ylabel('Nombre de fois')
plt.title('Nombre de fois que chaque type d\'action a été effectué')
plt.grid(False)

# Ajouter les valeurs au-dessus de chaque barre
for index, value in enumerate(df_collection_navigation['type_action'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center')

plt.show()
```

```python
# Créer un diagramme circulaire pour le navigateur utilisé
plt.figure(figsize=(8, 8))
df_collection_navigation['browser_used'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Répartition des navigateurs utilisés')
plt.show()
```

```python
df_collection_navigation['viewing_start_time'] = pd.to_datetime(df_collection_navigation['viewing_start_time'])
df_collection_navigation['viewing_end_time'] = pd.to_datetime(df_collection_navigation['viewing_end_time'])

# Calculer la durée de visualisation en minutes
df_collection_navigation['duration'] = (df_collection_navigation['viewing_end_time'] - df_collection_navigation['viewing_start_time']).dt.total_seconds() 
```

```python

# Tracer le diagramme de densité
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_collection_navigation['duration'], shade=False, color="#D64550")
plt.xlabel('Durée de visualisation (minutes)')
plt.ylabel('Densité')
plt.title('Diagramme de densité des durées de visualisation')
plt.xticks(range(0, 10, 1))  # Graduation de 0 à 40 avec des intervalles de 5
plt.show()
```
