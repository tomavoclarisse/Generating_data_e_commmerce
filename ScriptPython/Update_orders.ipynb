```python
import pandas as pd
```

```python
df_orders = pd.read_csv("dataset/orders.csv")
df_order_details = pd.read_csv("dataset/order_details.csv")
```

```python
def total_amount(df_orders, df_order_details): #somme amount_invoiced + total_amount de df_orders
    len_df_orders = len(df_orders)
    lst_total_amount = []
    lst_total_amount_invoiced = []
    for o in range(len_df_orders):
        sum_amount_invoiced =0
        sum_amount_net = 0
        total_amount_invoiced = 0
        total_amount = 0
        df_od_cu = df_order_details[df_order_details["Order_ID"]==df_orders["Order_ID"].iloc[o]]
        delivery_fees = df_orders["Total_amount"].iloc[o]
        len_df_od_cu = len(df_od_cu)
        for od in range(len_df_od_cu):
            sum_amount_invoiced = sum_amount_invoiced + df_od_cu["Amount_invoiced"].iloc[od]
            sum_amount_net = sum_amount_net + df_od_cu["Amount_net"].iloc[od]
    
        total_amount_invoiced = sum_amount_invoiced + delivery_fees
        total_amount = sum_amount_net + delivery_fees
        lst_total_amount.append(total_amount)
        lst_total_amount_invoiced.append(total_amount_invoiced)
    return lst_total_amount, lst_total_amount_invoiced
```

```python
lst_amount, lst_amount_invoiced = total_amount(df_orders, df_order_details)
```

```python
df_orders["Total_amount"] = lst_amount 
df_orders["Total_amount_invoiced"] = lst_amount_invoiced
```

```python

orders_csv = df_orders.to_csv('dataset\orders.csv', encoding='utf-8-sig', index=False)
```
