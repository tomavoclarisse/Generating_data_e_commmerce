## Génération des données Category

```python
import pymongo
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import string
from unidecode import unidecode
import warnings 
warnings.filterwarnings("ignore")
```

## Connexion à ma base de données mongodb

```python
# Connexion à MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/") 
database = client["dbProductElunea"] 
```

```python
# Accédons à la collection link_categories
col_categories = database["link_categories"]

# Lecture des données chargées
documents = col_categories.find()
data_cat = []
for document in documents:
    data_cat.append(document)
    
lst_all_cat = []
lst_descriptions = ["Des gammes de maquillage pour révéler votre beauté naturelle avec éclat.",
                    "Plongez dans notre univers olfactif et découvrez des parfums envoûtants qui captivent les sens.",
                    "Une sélection de soins de visage pour une peau éclatante et radieuse au quotidien.",
                    "Plongez dans un monde de bien-être avec notre gamme de soins pour le corps et le bain, pour une expérience sensorielle inoubliable.",
                    "Des collections de produits capillaires pour sublimer votre chevelure avec élégance."
                   ]
    
for i in range(len(data_cat)):
    cat_current = data_cat[i]
    dic_cat = {}
    if (cat_current["name_category"] =="SOINS VISAGE"):
        cat_name = "Soins du visage"
    else :
        cat_name = cat_current["name_category"]
    dic_cat["Category_name"] = cat_name.title().strip()
    dic_cat["Category_parent"] = ""
    dic_cat["Category_description"] = lst_descriptions[i]
    lst_all_cat.append(dic_cat)
    dic_cat = {}
    sub_cats = cat_current["sub_category"]
    for sub_cat in sub_cats : 
        
        dic_cat["Category_name"] = sub_cat["name_sub_category"].title().strip()
        dic_cat["Category_parent"] = cat_current ["name_category"].title().strip()
        dic_cat["Category_description"] = lst_descriptions[i]
        lst_all_cat.append(dic_cat)
        dic_cat = {}
        
            
# génération des cat_ID
list_Cat_ID = []
for i in range(1, len(lst_all_cat)+1):
    Cat_ID = f"Cat_{i:04d}"
    list_Cat_ID.append(Cat_ID)
    
    
# création du dataframe categories
lst_columns = ["Category_ID", "Category_name", "Category_parent", "Category_description"]
df_categories = pd.DataFrame(columns = lst_columns )
for cat in lst_all_cat : 
    df_categories = df_categories.append(cat, ignore_index=True)

df_categories["Category_ID"] = list_Cat_ID
```
